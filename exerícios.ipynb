```python
import numpy as np
import matplotlib.pyplot as plt

def euler_method(f, y0, t0, tf, h):
    n = int((tf - t0) / h)
    t = np.linspace(t0, tf, n + 1)
    y = np.zeros(n + 1)
    y[0] = y0
    for i in range(n):
        y[i + 1] = y[i] + h * f(t[i], y[i])
    return t, y

def calculate_errors(y_exact, y_approx):
    abs_error = np.abs(y_exact - y_approx)
    rel_error = np.abs(abs_error / y_exact)
    return abs_error, rel_error

def plot_solution(t_exact, y_exact, results, title):
    plt.figure(figsize=(10, 6))
    plt.plot(t_exact, y_exact, label="Solução Exata", color='black', linewidth=2)
    for (t, y, h) in results:
        plt.plot(t, y, marker='o', linestyle='--', label=f"Euler h={h}")
    plt.title(title)
    plt.xlabel("t")
    plt.ylabel("y(t)")
    plt.grid(True)
    plt.legend()
    plt.show()

# =====================
# 1. y' = y ; y(0) = 1
# =====================
f1 = lambda t, y: y
y_exact1 = lambda t: np.exp(t)

# =====================
# 2. y' = -y + t^2 + 1 ; y(0) = 2
# =====================
f2 = lambda t, y: -y + t**2 + 1
y_exact2 = lambda t: t**2 + 3 - 2*t - 1/np.exp(t)

# =====================
# 3. y' = cos(t) ; y(0) = 0
# =====================
f3 = lambda t, y: np.cos(t)
y_exact3 = lambda t: np.sin(t)

# Configurações
cases = [
    ("1. y' = y", f1, 1, 0, 2, y_exact1),
    ("2. y' = -y + t² + 1", f2, 2, 0, 2, y_exact2),
    ("3. y' = cos(t)", f3, 0, 0, 2 * np.pi, y_exact3)
]
hs = [0.1, 0.2, 0.5, 1, 2]

# Processa cada equação
for title, f, y0, t0, tf, y_exact_func in cases:
    t_exact = np.linspace(t0, tf, 1000)
    y_exact_vals = y_exact_func(t_exact)
    results = []

    print(f"\n=== {title} ===")
    for h in hs:
        t, y = euler_method(f, y0, t0, tf, h)
        y_exact_at_t = y_exact_func(t)
        abs_error, rel_error = calculate_errors(y_exact_at_t, y)
        results.append((t, y, h))

        print(f"\n--- h = {h} ---")
        for i in range(len(t)):
            print(f"t={t[i]:.2f}, y_euler={y[i]:.5f}, y_exact={y_exact_at_t[i]:.5f}, "
                  f"abs_error={abs_error[i]:.5f}, rel_error={rel_error[i]:.5f}")

    plot_solution(t_exact, y_exact_vals, results, title)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

def euler_melhorado(f, y0, t0, tf, h):
    n = int((tf - t0) / h)
    t = np.linspace(t0, tf, n + 1)
    y = np.zeros(n + 1)
    y[0] = y0
    for i in range(n):
        f1 = f(t[i], y[i])
        y_predict = y[i] + h * f1
        f2 = f(t[i+1], y_predict)
        y[i+1] = y[i] + (h / 2) * (f1 + f2)
    return t, y

def calculate_errors(y_exact, y_approx):
    abs_error = np.abs(y_exact - y_approx)
    rel_error = np.abs(abs_error / y_exact)
    return abs_error, rel_error

def plot_solution(t_exact, y_exact, results, title):
    plt.figure(figsize=(10, 6))
    plt.plot(t_exact, y_exact, label="Solução Exata", color='black', linewidth=2)
    for (t, y, h) in results:
        plt.plot(t, y, marker='o', linestyle='--', label=f"Euler Melhorado h={h}")
    plt.title(title)
    plt.xlabel("t")
    plt.ylabel("y(t)")
    plt.grid(True)
    plt.legend()
    plt.show()

    # Tabela de erros
    for t, y, h in results:
        y_exact_at_t = y_exact_func(t)
        abs_error, rel_error = calculate_errors(y_exact_at_t, y)

        df = pd.DataFrame({
            't': t,
            'Euler Melhorado y(t)': y,
            'Exato y(t)': y_exact_at_t,
            'Erro Absoluto': abs_error,
            'Erro Relativo (%)': rel_error * 100
        })

        print(f"\n### Tabela de Erros para h = {h} ###")
        display(df.round(5))

# =====================
# 1. y' = y ; y(0) = 1
# =====================
f1 = lambda t, y: y
y_exact1 = lambda t: np.exp(t)

# =====================
# 2. y' = -y + t^2 + 1 ; y(0) = 2
# =====================
f2 = lambda t, y: -y + t**2 + 1
y_exact2 = lambda t: t**2 + 3 - 2*t - 1/np.exp(t)

# =====================
# 3. y' = cos(t) ; y(0) = 0
# =====================
f3 = lambda t, y: np.cos(t)
y_exact3 = lambda t: np.sin(t)

# Configurações
cases = [
    ("1. y' = y", f1, 1, 0, 2, y_exact1),
    ("2. y' = -y + t² + 1", f2, 2, 0, 2, y_exact2),
    ("3. y' = cos(t)", f3, 0, 0, 2 * np.pi, y_exact3)
]
hs = [0.1, 0.2, 0.5, 1, 2]

# Processa cada equação
for title, f, y0, t0, tf, y_exact_func in cases:
    t_exact = np.linspace(t0, tf, 1000)
    y_exact_vals = y_exact_func(t_exact)
    results = []

    print(f"\n=== {title} ===")
    for h in hs:
        t, y = euler_melhorado(f, y0, t0, tf, h)
        results.append((t, y, h))

    plot_solution(t_exact, y_exact_vals, results, title)
```

# **Método de Euler para dois PVIs**


*   PVI 1: y′ = y, y(0) = 1 → Solução exata: y(t) = e^t

*   PVI 2: y′ = cos(t), y(0) = 0 → Solução exata: y(t) = sin(t)



```python
import numpy as np
import matplotlib.pyplot as plt # transformar isso em uma celula separada

# Método de Euler simples
def euler_method(f, y0, t0, tf, h):
    n = int((tf - t0) / h)
    t = np.linspace(t0, tf, n + 1)
    y = np.zeros(n + 1)
    y[0] = y0
    for i in range(n):
        y[i + 1] = y[i] + h * f(t[i], y[i])
    return t, y

# Função para exibir erros e gráficos
def analisar_euler(f, y_exata, y0, t0, tf, h, titulo):
    t, y_aprox = euler_method(f, y0, t0, tf, h)
    y_real = y_exata(t)
    erro_abs = np.abs(y_real - y_aprox)
    erro_rel = erro_abs / np.abs(y_real)

    # Gráfico da solução
    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.plot(t, y_real, 'k-', label="Solução Exata")
    plt.plot(t, y_aprox, 'bo--', label="Euler")
    plt.title(f"{titulo} - Solução")
    plt.xlabel("t"); plt.ylabel("y(t)"); plt.grid(); plt.legend()

    # Gráfico dos erros
    plt.subplot(1, 2, 2)
    plt.plot(t, erro_abs, 'r-', label="Erro Absoluto")
    plt.plot(t, erro_rel * 100, 'g--', label="Erro Relativo (%)")
    plt.title("Erros")
    plt.xlabel("t"); plt.grid(); plt.legend()
    plt.tight_layout()
    plt.show()

    # Print dos erros
    for i in range(len(t)):
        print(f"t = {t[i]:.2f} | y = {y_aprox[i]:.5f} | exato = {y_real[i]:.5f} | erro abs = {erro_abs[i]:.5f} | erro rel = {erro_rel[i]*100:.2f}%")

# PVI 1: y' = y
f1 = lambda t, y: y
y1_exata = lambda t: np.exp(t)
analisar_euler(f1, y1_exata, y0=1, t0=0, tf=2, h=0.2, titulo="PVI 1: y' = y")

# PVI 2: y' = cos(t)
f2 = lambda t, y: np.cos(t)
y2_exata = lambda t: np.sin(t)
analisar_euler(f2, y2_exata, y0=0, t0=0, tf=2*np.pi, h=0.2, titulo="PVI 2: y' = cos(t)")
```

```python
# Método de Euler Melhorado (Heun / RK2)
def euler_melhorado(f, y0, t0, tf, h):
    n = int((tf - t0) / h)
    t = np.linspace(t0, tf, n + 1)
    y = np.zeros(n + 1)
    y[0] = y0
    for i in range(n):
        k1 = f(t[i], y[i])
        k2 = f(t[i] + h, y[i] + h * k1)
        y[i + 1] = y[i] + (h / 2) * (k1 + k2)
    return t, y

# Função para análise com Euler Melhorado
def analisar_euler_melhorado(f, y_exata, y0, t0, tf, h, titulo):
    t, y_aprox = euler_melhorado(f, y0, t0, tf, h)
    y_real = y_exata(t)
    erro_abs = np.abs(y_real - y_aprox)
    erro_rel = erro_abs / np.abs(y_real)

    # Gráficos
    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.plot(t, y_real, 'k-', label="Solução Exata")
    plt.plot(t, y_aprox, 'bs--', label="Euler Melhorado")
    plt.title(f"{titulo} - Solução")
    plt.xlabel("t"); plt.ylabel("y(t)"); plt.grid(); plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(t, erro_abs, 'r-', label="Erro Absoluto")
    plt.plot(t, erro_rel * 100, 'g--', label="Erro Relativo (%)")
    plt.title("Erros")
    plt.xlabel("t"); plt.grid(); plt.legend()
    plt.tight_layout()
    plt.show()

    # Print dos erros
    for i in range(len(t)):
        print(f"t = {t[i]:.2f} | y = {y_aprox[i]:.5f} | exato = {y_real[i]:.5f} | erro abs = {erro_abs[i]:.5f} | erro rel = {erro_rel[i]*100:.2f}%")

# Reutilizando os mesmos PVIs
analisar_euler_melhorado(f1, y1_exata, y0=1, t0=0, tf=2, h=0.2, titulo="PVI 1: y' = y (RK2)")
analisar_euler_melhorado(f2, y2_exata, y0=0, t0=0, tf=2*np.pi, h=0.2, titulo="PVI 2: y' = cos(t) (RK2)")
```

```python
import matplotlib.pyplot as plt

# Parâmetros do modelo (PDF)
k = 1.0
gamma = 0.0
rho = 0.5
alpha = 0.75
sigma = 0.5

# Condições iniciais (presas > predadores)
t0 = 0.0
A0 = 2.0   # Presas
B0 = 1.5   # Predadores
h = 0.1
Nmax = 300  # Simulação por 30 unidades de tempo

# Modelo de Lotka-Volterra com Euler explícito
def funcao1(t, A, B):
    return k * A - gamma * A**2 - rho * A * B

def funcao2(t, A, B):
    return -alpha * B + sigma * A * B

# Vetores de solução
tt = [t0]
AA = [A0]
BB = [B0]

# Método de Euler
for i in range(Nmax):
    A = A0 + h * funcao1(t0, A0, B0)
    B = B0 + h * funcao2(t0, A0, B0)
    t0 += h
    AA.append(A)
    BB.append(B)
    tt.append(t0)
    A0, B0 = A, B

# Gráfico 1: Presas e predadores ao longo do tempo
plt.figure(figsize=(10,5))
plt.plot(tt, AA, 'g-', label='Presas (A)')
plt.plot(tt, BB, 'r-', label='Predadores (B)')
plt.xlabel("Tempo")
plt.ylabel("População")
plt.title("Modelo Presa-Predador (Exemplo PDF) - Euler")
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()

# Gráfico 2: Espaço de fases
plt.figure(figsize=(6,6))
plt.plot(AA, BB, 'k-', label='Órbita')
plt.plot(AA[0], BB[0], 'go', label='Início')
plt.plot(AA[-1], BB[-1], 'ro', label='Fim')
plt.xlabel("Presas (A)")
plt.ylabel("Predadores (B)")
plt.title("Espaço de Fases (A x B)")
plt.legend()
plt.grid()
plt.axis("equal")
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm

# MODELO PRESA-PREDADOR

# EDOs
def dxdt(x, y, b, h, epsilon, d):
    return b * x - h * x * y

def dydt(x, y, b, h, epsilon, d):
    return epsilon * h * x * y - d * y

def predador_presa(x0, y0, b, h, epsilon, d, delta_t, max_t):
    if x0 < y0:
        print("⚠️ Atenção: população inicial de predadores está maior que a de presas.") #aviso acionado se a população de predadores for maior do que de presas
    x, y, t = x0, y0, 0
    t_list, x_list, y_list = [t], [x], [y]
    for _ in tqdm(range(int(max_t / delta_t))):
        x += dxdt(x, y, b, h, epsilon, d) * delta_t
        y += dydt(x, y, b, h, epsilon, d) * delta_t
        t += delta_t
        t_list.append(t)
        x_list.append(x)
        y_list.append(y)
    return t_list, x_list, y_list

# Parâmetros
b = 1.0
h_ = 0.005
epsilon = 0.8
d = 0.6
x0, y0 = 100, 50  # população inicial com mais presas que predadores
delta_t = 0.001
max_t = 100

# simulação
t_sim, x_sim, y_sim = predador_presa(x0, y0, b, h_, epsilon, d, delta_t, max_t)

# Gráfico da evolução temporal
plt.plot(t_sim, x_sim, label="Presas (x)")
plt.plot(t_sim, y_sim, label="Predadores (y)")
plt.title("População x Tempo")
plt.xlabel("Tempo")
plt.ylabel("População")
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()

# Gráfico de órbita
plt.plot(x_sim, y_sim)
plt.title("Órbita no Espaço de Fases")
plt.xlabel("Presas (x)")
plt.ylabel("Predadores (y)")
plt.grid()
plt.axis('equal')
plt.tight_layout()
plt.show()


# exemplo

# Modelo com notação do PDF: A = presas, B = predadores
def modelo_pdf(t, U, k, gamma, rho, alpha, sigma):
    A, B = U
    dA = k * A - gamma * A**2 - rho * A * B
    dB = -alpha * B + sigma * A * B
    return [dA, dB]

# Método de Euler para sistemas
def euler_sistema(F, U0, t0, tf, h):
    n = int((tf - t0) / h)
    t = np.linspace(t0, tf, n + 1)
    U = np.zeros((n + 1, len(U0)))
    U[0] = U0
    for i in range(n):
        dU = F(t[i], U[i])
        U[i + 1] = U[i] + h * np.array(dU)
    return t, U

# Parâmetros do exemplo do PDF (γ = 0)
k = 1.0
gamma = 0.0
rho = 0.5
alpha = 0.75
sigma = 0.5

# Condições iniciais do exemplo - com mais presas que predadores
U0 = [2.0, 1.5]  # A = presas, B = predadores
t0, tf, h = 0, 30, 0.1

# Função com os parâmetros fixos
def F_pdf(t, U):
    return modelo_pdf(t, U, k, gamma, rho, alpha, sigma)

# simulação do exemplo
t1, U1 = euler_sistema(F_pdf, U0, t0, tf, h)

# Gráfico do exemplo do PDF
plt.plot(t1, U1[:, 0], label="Presas (A)", color='green')
plt.plot(t1, U1[:, 1], label="Predadores (B)", color='red')
plt.title("Exemplo do PDF - População vs Tempo")
plt.xlabel("Tempo")
plt.ylabel("População")
plt.grid()
plt.legend()
plt.tight_layout()
plt.show()

# Gráfico do espaço de fases (exemplo do PDF)
plt.plot(U1[:, 0], U1[:, 1])
plt.title("Exemplo do PDF - Órbita no Espaço de Fases")
plt.xlabel("Presas (A)")
plt.ylabel("Predadores (B)")
plt.grid()
plt.axis('equal')
plt.tight_layout()
plt.show()
```

Implementar o MTR (meotodo trapezios repetidos)
dados: f uma função contínua
a < b, a, b pertence aos reais
n é inteiro positivo


---



```python
#Aplicando a lógica do metodo de trapezios repetidos
def metodo_trapezios_repetidos(f, a, b, n):
    """
    Método dos Trapézios Repetidos para aproximar a integral de f no intervalo [a, b] com n subintervalos.

    Parâmetros:
    f -- função contínua a ser integrada
    a -- limite inferior (float)
    b -- limite superior (float)
    n -- número de subintervalos (int)

    Retorna:
    Aproximação numérica da integral de f no intervalo [a, b]
    """
    h = (b - a) / n
    soma = f(a) + f(b)

    for i in range(1, n): #utilizando o for como recomendado pelo professor
        xi = a + i * h
        soma += 2 * f(xi)

    return (h / 2) * soma
```

```python
#plotando o gráfico da equação
import numpy as np
import matplotlib.pyplot as plt

# Função f(x) = x(x - 10)
f = lambda x: x * (x - 10)

# Intervalo de integração
a, b = 0, 10

# Número de subintervalos (trapézios)
n = 8

# Nós e valores da função
x_nodes = np.linspace(a, b, n + 1)
y_nodes = f(x_nodes)

# Para a curva suave da função
x_fine = np.linspace(a, b, 500)
y_fine = f(x_fine)

# Figura
plt.figure(figsize=(10, 6))

# Curva da função
plt.plot(x_fine, y_fine, 'k-', linewidth=2, label='f(x) = x(x - 10)')

# Sombreamento dos trapézios
for i in range(n):
    x_trap = [x_nodes[i], x_nodes[i], x_nodes[i+1], x_nodes[i+1]]
    y_trap = [0, y_nodes[i], y_nodes[i+1], 0]
    plt.fill(x_trap, y_trap, color='orange', alpha=0.4, edgecolor='red', linewidth=1)

# Segmentos de reta (entre os nós)
for i in range(n):
    plt.plot([x_nodes[i], x_nodes[i+1]], [y_nodes[i], y_nodes[i+1]], 'r-', linewidth=2)

# Marcar os pontos amostrados
plt.scatter(x_nodes, y_nodes, color='blue', label='Nós')

# Estética
plt.title(f"Aproximação de f(x) = x(x - 10) com {n} Trapézios")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Função f(x) = sen(x)
f = lambda x: np.sin(x)

# Intervalo de integração
a = 0
b = np.pi

# Número de subintervalos (trapézios)
n = 8

# Nós e valores da função
x_nodes = np.linspace(a, b, n + 1)
y_nodes = f(x_nodes)

# Curva suave da função
x_fine = np.linspace(a, b, 500)
y_fine = f(x_fine)

# Figura
plt.figure(figsize=(10, 6))

# Curva da função
plt.plot(x_fine, y_fine, 'k-', linewidth=2, label='f(x) = sen(x)')

# Trapézios
for i in range(n):
    x_trap = [x_nodes[i], x_nodes[i], x_nodes[i+1], x_nodes[i+1]]
    y_trap = [0, y_nodes[i], y_nodes[i+1], 0]
    plt.fill(x_trap, y_trap, color='orange', alpha=0.4, edgecolor='red', linewidth=1)

# Segmentos de reta do método dos trapézios
for i in range(n):
    plt.plot([x_nodes[i], x_nodes[i+1]], [y_nodes[i], y_nodes[i+1]], 'r-', linewidth=2)

# Pontos amostrados
plt.scatter(x_nodes, y_nodes, color='blue', label='Nós')

# Estética
plt.title(f"Aproximação de f(x) = sen(x) com {n} Trapézios no intervalo [0, π]")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
