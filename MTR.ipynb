# Aplicando o metódo dos trapézios repetidos

> Implementar o MTR (meotodo trapezios repetidos) dados: f uma função contínua a < b, a, b pertence aos reais n é inteiro positivo

por Giulia Tafuri


```python
#Aplicando a lógica do metodo de trapezios repetidos
def metodo_trapezios_repetidos(f, a, b, n):
    """
    Método dos Trapézios Repetidos para aproximar a integral de f no intervalo [a, b] com n intervalos.

    Parâmetros:
    f -- função contínua
    a -- limite inferior
    b -- limite superior
    n -- número de subintervalos

    Retorna:
    Aproximação da integral
    """
    h = (b - a) / n
    soma = f(a) + f(b)

    for i in range(1, n): #utilizando o for como recomendado pelo professor
        xi = a + i * h
        soma += 2 * f(xi)

    return (h / 2) * soma
```

Aplicando a função f(x)= x(x-10)

```python
#exemplo de uso
import math

# Função a ser integrada
f = lambda x: x*(x-10)

# Intervalo e número de subintervalos
a = 0
b = 1
n = 100

# Cálculo da integral aproximada
aprox = metodo_trapezios_repetidos(f, a, b, n)
print(f"Aproximação da integral de x(x-10) de {a} a {b} com n={n}: {aprox:.6f}")
```

```python
#plotando o gráfico da equação
import numpy as np
import matplotlib.pyplot as plt

# Função f(x) = x(x - 10)
f = lambda x: x * (x - 10)

# Intervalo de integração
a, b = 0, 10

# Número de subintervalos (trapézios)
n = 8

# Nós e valores da função
x_nodes = np.linspace(a, b, n + 1)
y_nodes = f(x_nodes)

# Para a curva suave da função
x_fine = np.linspace(a, b, 500)
y_fine = f(x_fine)

# Figura
plt.figure(figsize=(10, 6))

# Curva da função
plt.plot(x_fine, y_fine, 'k-', linewidth=2, label='f(x) = x(x - 10)')

# Sombreamento dos trapézios
for i in range(n):
    x_trap = [x_nodes[i], x_nodes[i], x_nodes[i+1], x_nodes[i+1]]
    y_trap = [0, y_nodes[i], y_nodes[i+1], 0]
    plt.fill(x_trap, y_trap, color='orange', alpha=0.4, edgecolor='red', linewidth=1)

# Segmentos de reta (entre os nós)
for i in range(n):
    plt.plot([x_nodes[i], x_nodes[i+1]], [y_nodes[i], y_nodes[i+1]], 'r-', linewidth=2)

# Marcar os pontos amostrados
plt.scatter(x_nodes, y_nodes, color='blue', label='Nós')

# Estética
plt.title(f"Aproximação de f(x) = x(x - 10) com {n} Trapézios")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

Aplicando a função f(x) = sen(x)

```python
import numpy as np
import matplotlib.pyplot as plt

# Função f(x) = sen(x)
f = lambda x: np.sin(x)

# Intervalo de integração
a = 0
b = np.pi

# Número de subintervalos (trapézios)
n = 8

# Nós e valores da função
x_nodes = np.linspace(a, b, n + 1)
y_nodes = f(x_nodes)

# Curva suave da função
x_fine = np.linspace(a, b, 500)
y_fine = f(x_fine)

# Figura
plt.figure(figsize=(10, 6))

# Curva da função
plt.plot(x_fine, y_fine, 'k-', linewidth=2, label='f(x) = sen(x)')

# Trapézios
for i in range(n):
    x_trap = [x_nodes[i], x_nodes[i], x_nodes[i+1], x_nodes[i+1]]
    y_trap = [0, y_nodes[i], y_nodes[i+1], 0]
    plt.fill(x_trap, y_trap, color='orange', alpha=0.4, edgecolor='red', linewidth=1)

# Segmentos de reta do método dos trapézios
for i in range(n):
    plt.plot([x_nodes[i], x_nodes[i+1]], [y_nodes[i], y_nodes[i+1]], 'r-', linewidth=2)

# Pontos amostrados
plt.scatter(x_nodes, y_nodes, color='blue', label='Nós')

# Estética
plt.title(f"Aproximação de f(x) = sen(x) com {n} Trapézios no intervalo [0, π]")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
